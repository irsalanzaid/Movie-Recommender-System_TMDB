### Data pre-processing, cleaning

```python

df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
import ast
def convert(string):
    genre_list=[]
    for i in ast.literal_eval(string):                   #converting string into list
        genre_list.append(i['name'])
    return genre_list
```

```python
df['genres']=df['genres'].apply(convert)
print(df['genres'])
```

```python
#doing the same thing on keywords column
df['keywords'] = df['keywords'].apply(convert)
df['keywords']
```

```python
df.head()
```

```python
#need 3 names from cast column
def convert_cast(string):
    genre_list=[]
    flag = 0
    for i in ast.literal_eval(string): 
        if flag !=3:               #need first 3 names from cast
            genre_list.append(i['name'])
            flag+=1
        else:
            break
    return genre_list
```

```python
df['cast'] = df['cast'].apply(convert_cast)
df['cast']
```

```python
df['crew'][0]
```

```python
#need director name from crew column
def convert_crew(string):
    genre_list=[]
    for i in ast.literal_eval(string): 
        if i['job']=='Director':              
            genre_list.append(i['name'])
    return genre_list
```

```python
df['crew'] = df['crew'].apply(convert_crew)
df['crew']
```

```python
df.head(3)
```

```python
df['overview'][0]
```

```python
#converting string overview column in list
df['overview']= df['overview'].apply(lambda r:r.split())
df['overview']
```

```python
df.head(3)
```

```python
#Data transformation, removing spaces
df['genres']=df['genres'].apply(lambda r: [i.replace(" ","") for i in r])
df['keywords']=df['keywords'].apply(lambda r: [i.replace(" ","") for i in r])
df['cast']=df['cast'].apply(lambda r: [i.replace(" ","") for i in r])
df['crew']=df['crew'].apply(lambda r: [i.replace(" ","") for i in r])
```

```python
df.head()
```

```python
#Creating new columns and merging some columns into that new column
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
#We need only three columns now, i.e., id, title and tags
new_df = df[['id','title', 'tags']]
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#defining function to stem all texts
def stem(text):
    list = []
    for i in text.split():
        list.append(ps.stem(i))
    return " ".join(list)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
new_df['tags']
```

```python
new_df['tags'][0]
```

```python
#text vectorization
#stopwords removal
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
#calculating cosine distance to find similarities among vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
#Defining function to recommend
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
new_df
```

```python
recommend('The Dark Knight Rises')
```
